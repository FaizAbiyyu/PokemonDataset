```python
import pandas as pd

# Load the Excel file
file_path = 'Pokemon.csv'
pokemon_data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataframe to understand its structure
pokemon_data.head()
```

```python
# Display the first few rows of the dataframe to understand its structure
pokemon_data.tail()
```

```python
# Extracting Pokémon IDs
pokemon_ids = pokemon_data['#']
print(pokemon_ids)
```

```python
# Extracting Pokémon names
pokemon_names = pokemon_data['Name']
print(pokemon_names)
```

```python
# Extracting Pokémon types
pokemon_types = pokemon_data[['Type 1', 'Type 2']]
print(pokemon_types)
```

karena terdapat banyak tipe pokemon jadi saya menjadikannya dalam bentuk bar supaya dapat mengetahui berapa banyak tipe yang sama

```python
import matplotlib.pyplot as plt

# Combine both 'Type 1' and 'Type 2' into a single Series and count the occurrences of each type
pokemon_types_combined = pd.concat([pokemon_data['Type 1'], pokemon_data['Type 2']]).dropna()
type_counts = pokemon_types_combined.value_counts()

# Plotting
plt.figure(figsize=(14, 8))
type_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Pokémon Types')
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Extracting total stats for each Pokémon
pokemon_total_stats = pokemon_data['Total']
print(pokemon_total_stats)
```

```python
# Finding Pokémon with the highest total stats
highest_total_stats = pokemon_data.sort_values(by='Total', ascending=False).head(10)

plt.figure(figsize=(14, 8))
plt.barh(highest_total_stats['Name'], highest_total_stats['Total'], color='mediumpurple')
plt.xlabel('Total Stats')
plt.ylabel('Pokémon Name')
plt.title('Top 10 Pokémon by Total Stats')
plt.grid(axis='x', linestyle='--')
plt.show()
```

```python
# Finding Pokémon with the lowest total stats
lowest_total_stats = pokemon_data.sort_values(by='Total', ascending=False).tail(10)

plt.figure(figsize=(14, 8))
plt.barh(lowest_total_stats['Name'], lowest_total_stats['Total'], color='mediumpurple')
plt.xlabel('Total Stats')
plt.ylabel('Pokémon Name')
plt.title('Top 10 Pokémon by Total Stats')
plt.grid(axis='x', linestyle='--')
plt.show()
```

```python
# Finding Pokémon with the highest total HP
highest_total_hp = pokemon_data.sort_values(by='HP', ascending=False).head(10)

plt.figure(figsize=(14, 8))
plt.barh(highest_total_hp['Name'], highest_total_hp['Total'], color='blue')
plt.xlabel('Total HP')
plt.ylabel('Pokémon Name')
plt.title('Top 10 Pokémon by Total HP')
plt.grid(axis='x', linestyle='--')
plt.show()
```

```python
# Finding Pokémon with the lowest total HP
lowest_total_hp = pokemon_data.sort_values(by='HP', ascending=False).tail(10)

plt.figure(figsize=(14, 8))
plt.barh(lowest_total_hp['Name'], lowest_total_hp['Total'], color='blue')
plt.xlabel('Total HP')
plt.ylabel('Pokémon Name')
plt.title('Top 10 Pokémon by Total HP')
plt.grid(axis='x', linestyle='--')
plt.show()
```

```python
# Finding Pokémon with the highest total att, def, sp.att, sp.deff
pokemon_attack_defense = pokemon_data.sort_values(by=['Attack', 'Defense', 'Sp. Atk', 'Sp. Def'], ascending=False).head(10)[['Name','Attack', 'Defense', 'Sp. Atk', 'Sp. Def']]
pokemon_attack_defense
```

```python
# Finding Pokémon with the highest total speed
pokemon_attack_defense = pokemon_data.sort_values(by='Speed', ascending=False).head(10)[['Name','Speed']]
pokemon_attack_defense
```

```python
# Finding Pokémon with the lowest total speed
pokemon_attack_defense = pokemon_data.sort_values(by='Speed', ascending=False).tail(10)[['Name','Speed']]
pokemon_attack_defense
```

```python
# Extracting generation information for each Pokémon
pokemon_generation = pokemon_data[['Name','Generation']]
pokemon_generation
```

```python
# Extracting legendary status for each Pokémon
pokemon_legendary = pokemon_data[['Name','Legendary']]
pokemon_legendary
```

```python
# Creating visualizations for the distribution of Pokémon by generation and the count of legendary Pokémon

# Count of Pokémon by Generation
pokemon_by_generation = pokemon_data['Generation'].value_counts().sort_index()

# Count of Legendary Pokémon
legendary_count = pokemon_data['Legendary'].value_counts()

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Bar chart for Pokémon by Generation
ax[0].bar(pokemon_by_generation.index, pokemon_by_generation.values, color='lightblue')
ax[0].set_title('Distribution of Pokémon by Generation')
ax[0].set_xlabel('Generation')
ax[0].set_ylabel('Number of Pokémon')
ax[0].grid(axis='y', linestyle='--')

# Pie chart for Legendary Pokémon
ax[1].pie(legendary_count, labels=legendary_count.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightgray'])
ax[1].set_title('Proportion of Legendary Pokémon')

plt.tight_layout()
plt.show()
```
